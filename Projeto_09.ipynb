### Criação de nova coluna:

Nova coluna apenas com os nomes das ruas.


```python
enderecos = []
for endereco in restaurante['address']:
    endereco = endereco.split()[1:]
    endereco = ' '.join(endereco)
    enderecos.append(endereco)

restaurante['Street'] = enderecos
display(restaurante.head(10))
```

### Alterando valores boleanos:

Alterando os dados da coluna Chain (cadeias de estabelecimentos) para facilitar a visualização das analises de/para:
* True - Rede
* False - Único (Não participa de nehuma rede)

```python
restaurante.loc[restaurante['chain'] == True, 'chain'] = 'Rede'
restaurante.loc[restaurante['chain'] == False, 'chain'] = 'Unico'

display(restaurante)
```

### Renomeei o nome dos principais restaurantes:

```python
def rest(dados):
    if 'Subw'  in dados:
        return 'Subway'
    if 'starb' in dados:
        return 'Starbucks'
    if 'MCD' in dados:
        return 'MCDONALDS'
    if 'THE COFFEE BEAN & TEA LEAF' in dados:
        return 'THE COFFEE BEAN & TEA LEAF'
    if 'EL POLLO LOCO' in dados:
        return 'EL POLLO LOCO'
    if 'KFC'  in dados:
        return 'KFC'
    if 'KENTUCKY FRIED CHICKEN'  in dados:
        return 'KFC'
    if 'DOMINO'  in dados:
        return 'DOMINOS PIZZA'
    if 'TRIMANA'  in dados:
        return 'TRIMANA'
    if 'CARLS JR'  in dados:
        return 'CARLS JR'
    if 'YOGURTLAND'  in dados:
        return 'YOGURTLAND'
    if 'LOUISIANA FRIED CHICKEN'  in dados:
        return 'LOUISIANA FRIED CHICKEN'
    if 'CHIPOTLE'  in dados:
        return 'CHIPOTLE'
    if 'BASKIN ROBBINS'  in dados:
        return 'BASKIN ROBBINS'
    if 'HONG KONG E'  in dados:
        return 'HONG KONG EXPRESS'
    if 'WABA GRILL'  in dados:
        return 'WABA GRILL'
    if 'WINGSTOP'  in dados:
        return 'WINGSTOP'
    if 'PAPA JOHN'  in dados:
        return 'PAPA JOHNS PIZZA'
    if 'JERSEY MIKE'  in dados:
        return 'JERSEY MIKES SUB'
    else:
        return dados
```

```python
restaurante['object_name'] = restaurante['object_name'].apply(rest)
display(restaurante.head(5))
```

```python
restaurante['object_name'].unique().shape
```

```python
restaurante_renomeados= restaurante.groupby('object_name')['object_type'].count().sort_values(ascending=False).head(20)
display(restaurante_renomeados)
```

## <center><strong>ANÁLISE DOS DADOS:</strong>

### Quantidade de estabelecimentos por tipo:

```python
tipos_estabelecimentos = restaurante.groupby('object_type')['object_name'].count().reset_index()
tipos_estabelecimentos.columns = ['object_type', 'quantity']
tipos_estabelecimentos = tipos_estabelecimentos.sort_values(by=['quantity'], ascending=True)
display(tipos_estabelecimentos)
```

```python
fig = px.bar(tipos_estabelecimentos, x='object_type', y='quantity', 
             color_discrete_sequence=px.colors.qualitative.Pastel1,
             template='plotly_white',
             text='quantity',
             height=500
             
)

fig.update_layout(
    xaxis_title='Tipo de estabelecimento', # Etiqueta do eixo X
    yaxis_title=' '
)

fig.update_traces(textposition='outside', textfont_size=10)
fig.update_yaxes(showticklabels=False)


fig.update_layout(title={
    'text' : 'Quantidade de estabelecimentos por tipo:',
    'y': 0.9,
    'x': 0.5
})

fig.show()
```

Nesta amostra os restaurantes são a maioria dos estabelecimentos, seguidos de Fast Food e cafés.
